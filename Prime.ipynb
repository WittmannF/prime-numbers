```
from math import sqrt
def notPrime(number):
    root=int(sqrt(number))
    for i in range(2, root+1):
        if number%i==0:
            return i
    return -1
```

```
START=1
END=100000000000000000000000000000000

all_numbers=[]
not_prime=[]
for n in range(START, END+1, 2):
    #print(f'{n} {notPrime(n)}')
    all_numbers.append(n)
    not_prime.append(notPrime(n))

```
